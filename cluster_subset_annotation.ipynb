```python
import os
import numpy as np
import pandas as pd

wd='/Users/juyoung/Desktop/immunotherapy_jy/singlecell_jy/cellmarker/'+project
print(wd)
print(cluster_csv_num)

#정리된 파일 저장할 곳 만들기
def output_make(num):
    nwd=wd+'/cluster'+str(num)
    os.makedirs(nwd)
    os.makedirs(nwd+'/output/')
    
#각 gene별로 cell type freq 계산 -> max인 cell type 선택
def df_freq(df,gene):
    freq=pd.DataFrame(df[['Cell Type']].value_counts())
    freq.columns=[gene]
    condition=freq[gene]==freq[gene].max()
    freq=freq[condition]
    #print(freq)
    return freq

#cluster안에 있는 30개 유전자의 df_freq함수 돌린 df들을 하나로 합침 -> final freq파일 만듬
def df_join(old,new):
    old=old.join(new, how='outer')
    old=old.replace(np.nan,0)
    
    return old

#sum값을 만듬
def df_max(old):
    s=pd.DataFrame(old.sum(axis=1))
    s.columns=['sum']
    
    old=old.join(s)
    return old

#최종 파일에서 sum이 max가 되는 모든 cell type을 가져와서 print
#최종파일 저장함 
def df_celltype(df,nwd):
    M=int(df['sum'].max())
    result=list(df[df['sum'] == M ].index)
    
    with pd.ExcelWriter(nwd+"/output/output.xlsx") as writer:
        df.to_excel(writer) 
   
    return (result)
    
    


for cluster_num in cluster_csv_num:
    nwd=wd+'/cluster'+str(cluster_num)+'/'
    gene_list=os.listdir(nwd)

    #output_make(cluster_num)
    
    if'.DS_Store'  in gene_list: gene_list.remove('.DS_Store')
    if 'output'  in gene_list: gene_list.remove('output')
    
    for i in range(len(gene_list)):  
        file=gene_list[i]
        gene=file[:-5]
        
        g=pd.read_excel(nwd+file)#csv파일로 저장되면 처음부터 column
        if g.iloc[0,0] == 'Species':  #xlsx로 저장되면 처음이 species 였음
            g=pd.read_excel(nwd+file,header=1)#그래서 header 지정해서 한번더 읽음
    
        if i==0:
            freq=df_freq(g,gene)
        else:
            gd=df_freq(g,gene)
            freq=df_join(freq,gd)
   
    freq=df_max(freq)
    #print('cell type of' ,cluster[1:-1],df_celltype(freq))
    print(df_celltype(freq,nwd))
   
    
```
